```python
data=d.drop("Unnamed: 0",axis=1)
data.head()
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(c,annot=True)
```

```python
X=data.drop("quality",axis=1)
Y=data["quality"]
X.shape , Y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
model=LogisticRegression(penalty='l2',solver='liblinear').fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
print("Accuracry score is :",accuracy_score(y_test,y_pred))
print("Precision score is :",precision_score(y_test,y_pred, average=None))
print("Recall score is :",recall_score(y_test,y_pred,average=None))
```
